```python
#Importing regressioon models and creating objects and making a list of those objects

#Linear Regression
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

#K-nearest-neighbours
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor(n_neighbors = 5)

#Decision Tree
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()

#Random Forest
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()


#A list of all the regressor models
models_list = [lr, knn, dt, rf]
```

```python
#Importing r2 score to check accuracy
from sklearn.metrics import r2_score


#A for loop that will iterate through models list, create and train a each model and print prediction Score
for model in models_list:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    score = r2_score(y_test, y_pred)
    print(f'For {model} model your accuracy is {score*100}%')
```
